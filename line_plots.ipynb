```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("train_HK6lq50.csv", encoding='mbcs')
```

```python
print(df.head())
```

```python
plt.plot(df["age"],df["difficulty_level"])
```

```python
x=df.pivot_table(index="program_type",values="age",aggfunc="mean").plot(kind="line")
```

```python
y=df.pivot_table(index="program_type",values=["age","trainee_engagement_rating"],aggfunc="mean").plot(kind="line")
```

```python
df.pivot_table(index = 'gender',values = ['age'],aggfunc = 'mean').plot(kind='line', color='yellow')

df.pivot_table(index = 'education',values = ['age'],aggfunc = 'mean').plot(kind='line', color='blue')


df.pivot_table(index = 'city_tier',values = ['age'],aggfunc = 'mean').plot(kind='line', color='green')



df.pivot_table(index = 'difficulty_level',values = ['total_programs_enrolled'],aggfunc = 'mean').plot(kind='line', color='orange')

plt.show()
```

```python
plt.rcParams['figure.figsize']=(12,5)
plt.subplot(221)
x=df.pivot_table(index = 'gender',values = ['age'],aggfunc = 'mean')
plt.plot(x);
plt.ylabel("age")
plt.xlabel("gender")

plt.subplot(222)
y=df.pivot_table(index = 'education',values = ['age'],aggfunc = 'mean')
plt.plot(y,color="yellow");
plt.ylabel("age")
plt.xlabel("education")

plt.subplot(223)
z=df.pivot_table(index = 'city_tier',values = ['age'],aggfunc = 'mean')
plt.plot(z,color="grey");
plt.ylabel("age")
plt.xlabel("city_tier")

plt.subplot(224)
a=df.pivot_table(index = 'difficulty_level',values = ['total_programs_enrolled'],aggfunc = 'mean')
plt.plot(a,color="purple");
plt.ylabel("total_programs_enrolled")
plt.xlabel("difficulty_level")

plt.show()
```

```python
plt.rcParams['figure.figsize']=(12,5)
plt.subplot(221)
x=df.pivot_table(index = 'gender',values = ['age'],aggfunc = 'mean')
plt.plot(x,"--");
plt.ylabel("age")
plt.xlabel("gender")

plt.subplot(122)
y=df.pivot_table(index = 'education',values = ['age'],aggfunc = 'mean')
plt.plot(y,'-+',color="black");
plt.ylabel("age")
plt.xlabel("education")

plt.subplot(223)
z=df.pivot_table(index = 'city_tier',values = ['age'],aggfunc = 'mean')
plt.plot(z,"^",color="grey");
plt.ylabel("age")
plt.xlabel("city_tier")

plt.show()
```
